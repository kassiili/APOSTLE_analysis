# Tracing all halos

```python
sim_id = "CDM_V1_LR"
snap = snapshot_obj.Snapshot(sim_id, 127)
```

```python
%time tracer = trace_halo.trace_all(snap, stop=115)
```

```python
print(tracer.size)
gns = snap.get_subhalos("GroupNumber")
sgns = snap.get_subhalos("SubGroupNumber")
for i,l in enumerate(tracer):
    out = "({},{})".format(gns[i],sgns[i])
    for elem in l[1:]:
        sid, idx = elem
        snapi = snapshot_obj.Snapshot(sim_id,sid)
        gnsi = snapi.get_subhalos("GroupNumber")
        sgnsi = snapi.get_subhalos("SubGroupNumber")
        if idx is None:
            out += " -> None"
        else:
            out += " -> ({},{})".format(gnsi[idx],sgnsi[idx])
    print(out)
```

```python
for snap_id, matches in tracer.items():
    print("{} : {}".format(snap_id,snap_id-1))
    snap1 = snapshot_obj.Snapshot(sim_id, snap_id)
    gns1 = snap1.get_subhalos("GroupNumber")
    sgns1 = snap1.get_subhalos("SubGroupNumber")
    snap2 = snapshot_obj.Snapshot(sim_id, snap_id-1)
    gns2 = snap2.get_subhalos("GroupNumber")
    sgns2 = snap2.get_subhalos("SubGroupNumber")
    for idx1, idx2 in enumerate(matches):
        if idx2 is None:
            print("({},{}) : ({},{})".format(gns1[idx1],sgns1[idx1],-1,-1))
        else:
            print("({},{}) : ({},{})".format(gns1[idx1],sgns1[idx1],gns2[idx2],sgns2[idx2]))
```
